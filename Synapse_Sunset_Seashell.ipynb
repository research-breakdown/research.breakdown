```python
from manim import *
```

```python
%%manim scene_2

class SpiralPolygon(VGroup):
  def __init__(self, polygon:RegularPolygon, n=150, prop=0.2, **kwargs):
    super().__init__(**kwargs)
    vertices = polygon.get_vertices()
    n_vertices = len(vertices)
    lines = [
        Line(vertices[i], vertices[(i+1)%n_vertices])
        for i in range(n_vertices)
    ]
    tmp_lines = lines[1:]
    for _ in range(n):
      last_line = tmp_lines[-1]
      first_line = tmp_lines[0]
      next_line = Line(
          last_line.get_end(),
          first_line.point_from_proportion(prop)
    )
      tmp_lines.pop(0)
      tmp_lines.append(next_line)
      self.add(next_line)

class scene_2(Scene):
  def construct(self):

    polygon = RegularPolygon(6).scale(3)
    center = polygon.get_center()
    vertices = polygon.get_vertices()
    n_vertices = len(vertices)

    lines = [
        Line(vertices[i], vertices[(i+1)%n_vertices])
        for i in range(n_vertices)
    ]
    mid_vertex = [l.point_from_proportion(0.618) for l in lines]
    overall_shape = VGroup(*[
        Polygon(
            mid_vertex[i], vertices[i], mid_vertex[i-1], center
        )
        for i in range(n_vertices)
    ]).set_stroke(width=3).set_color_by_gradient(ORANGE, PINK)
    inside_spiral = VGroup(*[
        SpiralPolygon(_p)
        for _p in overall_shape
    ]).set_color_by_gradient(PINK, YELLOW)
#-----------------------------------------------------------------------------#

    # final = VGroup(overall_shape, inside_spiral) #suspend this line to view animated version
    # self.add(final) #suspend this line to view animated version
    self.play(Create(overall_shape), Create(inside_spiral), run_time=5) #UNsuspend this line to view animated version
    self.wait(2)
```
